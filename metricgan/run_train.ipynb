# MetricGAN Training starten

Dieses Notebook startet `train.py` aus dem Ordner `metricgan` mit einer Hyperparameter-Datei.

```python
from pathlib import Path
import subprocess
import sys

project_dir = Path.cwd()
script_path = project_dir / 'train.py'
hparams_file = project_dir / 'hparams' / 'train.yaml'

# Datensatzpfad
data_folder = '.../data'

# Ergebnisordner
output_folder = './results'

if not script_path.exists():
    raise FileNotFoundError(f'Nicht gefunden: {script_path}')
if not hparams_file.exists():
    raise FileNotFoundError(f'Nicht gefunden: {hparams_file}')

cmd = [
    sys.executable,
    str(script_path),
    str(hparams_file),
    '--data_folder',
    data_folder,
    '--output_folder',
    output_folder,
]

print('Starte:', ' '.join(cmd))
subprocess.run(cmd, check=True, cwd=project_dir)
```
